# Automated Road Condition Assessment 

## Using an Integration of ArcGIS and TensorFlow to automate the assessment of road pavement conditions using collected video and images from driving inspections. 

This process extracts frames from a dashcam video as inputs to an object detector trained to detect different types of pavement condition damage and deposits the information on an ArcGIS Feature Service for consumption through mapping applications and information products.

### 4. Create General Helper Functions

Image conversion utility for TensorFlow operations

```python
def load_image_into_numpy_array(image):
    (im_width, im_height) = image.size
    return np.array(image.getdata()).reshape(
        (im_height, im_width, 3)).astype(np.uint8)
```

Fault counter utility

```python
def fault_counter(classes_arr, scores_arr, label_reference_dict, score_thresh=0.3):
    # Process the numpy array of classes from the model
    stacked_arr = np.stack((classes_arr, scores_arr), axis=-1)
    # Convert to pandas dataframe for easier querying
    detection_df = pd.DataFrame(stacked_arr)
    
    # Retrieve total count of each fault
    detected_D00 = detection_df[(detection_df[0] == 1.0) & (detection_df[1] > score_thresh)]
    detected_D01 = detection_df[(detection_df[0] == 2.0) & (detection_df[1] > score_thresh)]
    detected_D10 = detection_df[(detection_df[0] == 3.0) & (detection_df[1] > score_thresh)]
    detected_D11 = detection_df[(detection_df[0] == 4.0) & (detection_df[1] > score_thresh)]
    detected_D20 = detection_df[(detection_df[0] == 5.0) & (detection_df[1] > score_thresh)]
    detected_D40 = detection_df[(detection_df[0] == 6.0) & (detection_df[1] > score_thresh)]
    detected_D43 = detection_df[(detection_df[0] == 7.0) & (detection_df[1] > score_thresh)]
    detected_D44 = detection_df[(detection_df[0] == 8.0) & (detection_df[1] > score_thresh)]
    
    lon_cracks_count = len(detected_D00) + len(detected_D01)
    lat_cracks_count = len(detected_D10) + len(detected_D11)
    allig_cracks_count = len(detected_D20)
    othcorr_count = len(detected_D40) + len(detected_D43) + len(detected_D44)
    
    return lon_cracks_count, lat_cracks_count, allig_cracks_count, othcorr_count
```

Extract image from video using frame timestamp

```python
def extract_frame(video_path, time_stamp, out_image_path):
    clip = mpy.VideoFileClip(video_path)
    clip.save_frame(out_image_path, time_stamp)
    return out_images_path
```

Extract images for duration of video using interval input

```python
def extract_frames_for_video_at_intervals(video_path, interval, out_directory_path):
    clip = mpy.VideoFileClip(video_path)
    iterations = int(clip.duration / interval)
    time_counter = 0
    for iteration in range(iterations):
        time_counter += interval
        print("Exporting image at {0} seconds...".format(str(time_counter)))
        file_name = "Video_{0}.png".format(str(time_counter))
        out_path = os.path.join(out_directory_path, file_name)
        clip.save_frame(out_path, (time_counter))
```

Utility to run pavement condition inference on a single image

```python
def run_condition_inference_on_image(image_path, image_size=(12, 8), save_image=True, saved_image_path="testing\\test.png"):
    with detection_graph.as_default():
        with tf.Session(graph=detection_graph) as sess:
            # Definite input and output Tensors for detection_graph
            image_tensor = detection_graph.get_tensor_by_name('image_tensor:0')
            # Each box represents a part of the image where a particular object was detected.
            detection_boxes = detection_graph.get_tensor_by_name('detection_boxes:0')
            # Each score represent how level of confidence for each of the objects.
            # Score is shown on the result image, together with the class label.
            detection_scores = detection_graph.get_tensor_by_name('detection_scores:0')
            detection_classes = detection_graph.get_tensor_by_name('detection_classes:0')
            num_detections = detection_graph.get_tensor_by_name('num_detections:0')
            image = Image.open(image_path)
            area = (650, 720, 1270, 950)
            image = image.crop(area)
            # the array based representation of the image will be used later in order to prepare the
            # result image with boxes and labels on it.
            image_np = load_image_into_numpy_array(image)
            # Expand dimensions since the model expects images to have shape: [1, None, None, 3]
            image_np_expanded = np.expand_dims(image_np, axis=0)
            # Actual detection.
            (boxes, scores, classes, num) = sess.run(
              [detection_boxes, detection_scores, detection_classes, num_detections],
              feed_dict={image_tensor: image_np_expanded})
            # Visualization of the results of a detection.
            vis_util.visualize_boxes_and_labels_on_image_array(
                                                              image_np,
                                                              np.squeeze(boxes),
                                                              np.squeeze(classes).astype(np.int32),
                                                              np.squeeze(scores),
                                                              category_index,
                                                              min_score_thresh=0.3,
                                                              use_normalized_coordinates=True,
                                                              line_thickness=8)
            plt.figure(figsize=image_size)
            plt.imshow(image_np)
            if save_image:
                scipy.misc.imsave(saved_image_path, image_np)
```
